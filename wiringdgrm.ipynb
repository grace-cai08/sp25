```python
import pandas as pd
```

```python
connections = pd.read_csv('connections_princeton.csv.gz')
connections
```

```python
baker_cells = pd.read_csv('Auditory neurons Lockwood et al 2026 - Sheet1.csv')
baker_cells
```

```python
celltypes = baker_cells[['Cell type (Lockwood et al)', 'Codex June 2024', 'NT2026 (FAFB and BANC)', 'layer_mean', 'Across-hemi_layer_mean']]
celltypes
```

```python
lockwood_ids = celltypes['Codex June 2024']
connections_filtered = connections[(connections['pre_root_id'].isin(lockwood_ids)) & (connections['post_root_id'].isin(lockwood_ids))]
connections_filtered
```

```python
firstmerge = connections_filtered.merge(celltypes, how='left', left_on='pre_root_id', right_on='Codex June 2024').drop(columns=['Codex June 2024']).rename(columns={'Cell type (Lockwood et al)':'type_pre'})
conns = firstmerge.merge(celltypes, how='left', left_on='post_root_id', right_on='Codex June 2024').drop(columns=['Codex June 2024']).rename(columns={'Cell type (Lockwood et al)':'type_post'})
conns = conns.rename(columns={'NT2026 (FAFB and BANC)_x':'nt_type_banc', 'layer_mean_x':'layer_mean_pre', 'Across-hemi_layer_mean_x':'across_hemi_layer_pre', 'layer_mean_y':'layer_mean_post', 'Across-hemi_layer_mean_y':'across_hemi_layer_post'}).drop(columns=['NT2026 (FAFB and BANC)_y'])
conns = conns[['pre_root_id', 'post_root_id', 'syn_count', 'nt_type_banc', 'type_pre', 'type_post', 'layer_mean_pre', 'layer_mean_post', 'across_hemi_layer_pre', 'across_hemi_layer_post']]
conns
```

```python
g_all = conns.copy().rename(columns={'pre_root_id':'Source', 'post_root_id':'Target', 'syn_count':'Weight'})
```

```python
g_all.to_csv('g_all.csv', index=False)
```

```python
celltypes
```

```python
g_all_nodelist = celltypes.copy()
g_all_nodelist = g_all_nodelist.rename(columns={'Codex June 2024':'Id', 'Cell type (Lockwood et al)':'Label', 'layer_mean':'layer_mean[z]'})
g_all_nodelist = g_all_nodelist[['Id', 'Label', 'NT2026 (FAFB and BANC)', 'layer_mean[z]', 'Across-hemi_layer_mean']]
g_all_nodelist
```

```python
g_all_nodelist.to_csv('g_all_nodelist.csv', index=False)
```

```python
conns_bytype = conns.groupby(['type_pre', 'type_post', 'nt_type_banc'], as_index=False)[['syn_count']].sum().sort_values(by='syn_count', ascending=False)
conns_bytype
```

```python
g1 = conns_bytype.rename(columns={'type_pre':'Source', 'type_post':'Target', 'syn_count':'Weight'})
g1
```

```python
g2 = conns.groupby(['type_pre', 'type_post', 'nt_type_banc', 'layer_mean_pre', 'layer_mean_post'], as_index=False)[['syn_count']].sum().sort_values(by='syn_count', ascending=False)
g2 = g2.rename(columns={'type_pre':'Source', 'type_post':'Target', 'syn_count':'Weight'})
g2
```

```python
g2.to_csv('g2.csv', index=False)
```

```python
g3 = conns.groupby(['type_pre', 'type_post', 'nt_type_banc', 'across_hemi_layer_pre', 'across_hemi_layer_post'], as_index=False, dropna=False)[['syn_count']].sum().sort_values(by=['syn_count'], ascending=False)
g3 = g3.rename(columns={'type_pre':'Source', 'type_post':'Target', 'syn_count':'Weight'})
g3 = g3[['Source', 'Target', 'Weight', 'nt_type_banc', 'across_hemi_layer_pre', 'across_hemi_layer_post']]
g3
```

```python
g3.to_csv('g3.csv', index=False)
```

```python
g3_nodelist = celltypes.copy()
g3_nodelist = g3_nodelist.groupby(['Cell type (Lockwood et al)', 'NT2026 (FAFB and BANC)'], as_index=False)[['Across-hemi_layer_mean']].mean()
g3_nodelist = g3_nodelist.rename(columns={'Cell type (Lockwood et al)':'Id', 'NT2026 (FAFB and BANC)':'nt_type_banc', 'Across-hemi_layer_mean':'layer[z]'})
g3_nodelist
```

```python
g3_nodelist.to_csv('g3_nodelist.csv', index=False)
```
